```python
#Initial imports

from pyfinviz.quote import Quote #pyfinviz was installed using pip install.  The library allows the user to scrape data from www.finviz.com 
import pandas as pd
```

```python
#The available ratios for five comparable stocks were pulled and put into a pandas dataframe
#COIN = Coinbase

quote = Quote(ticker='COIN')
coin_ratios = pd.DataFrame(quote.fundamental_df)
coin_ratios
```

```python
#SNOW = Snowflake Inc.

quote = Quote(ticker='SNOW')
snow_ratios = pd.DataFrame(quote.fundamental_df)
snow_ratios
```

```python
#CRM = Salesforce

quote = Quote(ticker='CRM')
crm_ratios = pd.DataFrame(quote.fundamental_df)
crm_ratios
```

```python
#INTU = Intuit

quote = Quote(ticker='INTU')
intu_ratios = pd.DataFrame(quote.fundamental_df)
intu_ratios
```

```python
#NOW = Servicenow

quote = Quote(ticker='NOW')
now_ratios = pd.DataFrame(quote.fundamental_df)
now_ratios
```

```python
#The five dataframes for the comparable stocks were concatenated

condensed_df = pd.concat([coin_ratios, snow_ratios, crm_ratios, intu_ratios, now_ratios], axis=0)
condensed_df
```

```python
#A row was added to the above dataframe to list the ticker symbol for the stock

stock_index = ['COIN', 'SNOW', 'CRM', 'INTU', 'NOW']
condensed_df['Stock'] = stock_index
condensed_df
```

```python
#The row added above was set as the index for the dataframe

condensed_df.set_index('Stock', inplace=True)
condensed_df
```

```python
#The dataframe was cleaned to keep only the relevant financial ratios

clean_df = condensed_df[['P/E', 'EPS (ttm)', 'Forward P/E', 'P/S', 'P/B', 'ROI']]
clean_df
```

```python
#The cleaned dataframe was exported as a csv

clean_df.to_csv('voyager_ratios.csv')
```
